```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset =pd.read_csv(r"C:\Users\DELL\Downloads\archive (7)\Iris.csv")
dataset.head(3)

```

```python
sns.pairplot(data=dataset)
plt.show()
```

```python
import scipy.cluster.hierarchy as sc
```

```python
sc.dendrogram(sc.linkage(dataset,method='single', metric='euclidean'))
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
ac=AgglomerativeClustering(n_clusters=2,linkage='single')
dataset['Predict']=ac.fit_predict(dataset)
dataset.head(2)
```

```python
sns.pairplot(data=dataset,hue='Predict')
plt.show()
```
